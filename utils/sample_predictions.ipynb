```python
import pandas as pd
```

Gemma Predictions Samples 

```python
predictions_df = pd.read_csv("../logs/outputs/gemma/test_predictions_gemma-02.csv")
predictions_df.head()
```

```python
# sample 4 predictions from the predictions dataframe with labels 0 with confidence spread across 0 to 0.5 and 1 misclassified sample
benign_samples = predictions_df[predictions_df["label"] == 0]
benign_samples = benign_samples.sort_values(by="prediction_probability", ascending=True)
benign_samples = benign_samples[benign_samples["prediction_probability"] <= 0.5]
benign_samples = benign_samples.sample(4, random_state=42)
# sample 1 misclassified sample with label 0 and confidence above 0.5
misclassified_sample = predictions_df[(predictions_df["label"] == 0) & (predictions_df["prediction_probability"] > 0.5)]
misclassified_sample = misclassified_sample.sample(1, random_state=42)
# combine the two samples into one dataframe
benign_samples = pd.concat([benign_samples, misclassified_sample])

# sample 5 predictions from the predictions dataframe with labels 1 with confidence spread across 0.5 to 1
malicious_samples = predictions_df[predictions_df["label"] == 1]
malicious_samples = malicious_samples.sort_values(by="prediction_probability", ascending=False)
malicious_samples = malicious_samples[malicious_samples["prediction_probability"] >= 0.5]
malicious_samples = malicious_samples.sample(4, random_state=42)
# sample 1 misclassified sample with label 1 and confidence below 0.5
misclassified_sample = predictions_df[(predictions_df["label"] == 1) & (predictions_df["prediction_probability"] < 0.5)]
misclassified_sample = misclassified_sample.sample(1, random_state=42)
malicious_samples = pd.concat([malicious_samples, misclassified_sample])
# combine the two samples into one dataframe
sample_predictions = pd.concat([benign_samples, malicious_samples])

# remove columns category, conv_type, and source
sample_predictions = sample_predictions.drop(columns=["category", "conv_type", "source"]) 

# reorder columns to have messages, raw_text, label, prediction_probability
sample_predictions = sample_predictions[["messages", "raw_text", "label", "predicted_label", "prediction_probability"]]
print(sample_predictions)
```

```python
# save the sample predictions to a csv file
sample_predictions.to_csv("../logs/outputs/sample_predictions_gemma.csv", index=False)
```
